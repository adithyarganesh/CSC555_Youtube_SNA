```python
import networkx as nx
```

```python
z = nx.pagerank(G, alpha = 0.85, max_iter = 6)
```

```python
z
```

```python
f =  open("youtube_community.txt", "r")
```

```python
f
```

```python
lis = []
t = []
for i in f:
    pr = 0
    temp = i.split(",")
    for j in temp:
        pr += z[int(j)]
    lis.append(pr/len(temp))
    t.append(len(temp))
    
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot([x for x in range(len(lis))], lis)
```
